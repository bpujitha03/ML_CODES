```python
# Outlier Analysis
fig, axs = plt.subplots(3, figsize = (5,5))
plt1 = sns.boxplot(df['GRE Score'], ax = axs[0])
plt2 = sns.boxplot(df['CGPA'], ax = axs[1])
plt3 = sns.boxplot(df['TOEFL Score'], ax = axs[2])
plt.tight_layout()
```

```python
#Exploratory Data Analysis
sns.boxplot(df['CGPA'])
plt.show()
```

```python
# Let's see how Sales are related with other variables using scatter plot.
sns.pairplot(df, x_vars=['CGPA', 'GRE Score', 'TOEFL Score'], y_vars='Research', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
sns.jointplot(x="CGPA",y="LOR",data=df,kind="reg")
plt.show()
```
